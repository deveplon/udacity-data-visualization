# Overplotting, Transparency, and Jitter

The visualization will suffer from overplotting, where the high amount of overlap in points makes it difficult to see the actual relationship between the plotted variables.

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# Import data
fuel_econ = pd.read_csv('./data/fuel-econ.csv')
print(fuel_econ.shape)
fuel_econ.head(10)
```

###### Jitter - Randomly add/subtract a small value to each data point

```python
# Initialize plot figure
plt.figure(figsize=[20, 5])

# Left: Defualt scatter
plt.subplot(1, 2, 1)
sb.regplot(data=fuel_econ, x='year', y='comb', truncate=False, fit_reg=False)

# Right: Scatter with Jitter
# The x_jitter will make each x value will be adjusted randomly by +/-0.3
plt.subplot(1, 2, 2)
sb.regplot(data=fuel_econ, x='year', y='comb', truncate=False, fit_reg=False, x_jitter=0.3);
```

###### Jitter - Plot with both Jitter and Transparency

```python
sb.regplot(
    data = fuel_econ,
    x = 'year',
    y = 'comb',
    truncate=False,
    x_jitter=0.3,
    scatter_kws={'alpha':1/20}
);

plt.scatter(data = fuel_econ, x = 'year', y = 'comb', alpha=1/20);
```
